```python
import altair as alt
import pandas as pd
import numpy as np
```

```python
from vega_datasets import data

cars = data.cars()
cars.head()
```

5.1 Реализовать графики в Altair с использованием следующих ключей:

Метки: mark_text(), mark_rect().

Столбцы данных: , 'Weight_in_lbs', 'Year'.

Кодировки: x, y, color.

```python
bars = alt.Chart(cars).mark_rect().encode(
    x=alt.X('Year:O', timeUnit='year'),
    y=alt.Y('mean(Weight_in_lbs):Q'),
    color='mean(Weight_in_lbs):Q'
)
text = bars.mark_text(
    align='center',
    baseline='bottom',
    dy=-5
).encode(
    text=alt.Text('mean(Weight_in_lbs):Q', format='.1f')
)

chart = (bars + text).interactive()
chart
```

5.2. Используя данные об автомобилях, создайте диаграмму рассеяния (*scatter-plot*), на которой *размер* (*size*) точек становится больше при наведении на них курсора. **Данный график построен по моим значениям, Он не информативен, поэтому далее был построен график с другими значениями**

```python
hover = alt.selection_single(
    on='mouseover',
    fields=['Name'],
    nearest=True,
    empty='none'
)
scatter_chart = alt.Chart(cars).mark_circle(size=60).encode(
    x='Year',
    y='Weight_in_lbs:Q',
    color='Origin:N',
    tooltip=['Name', 'Year', 'Weight_in_lbs']
).add_selection(
    hover
).encode(
    size=alt.condition(hover, alt.value(200), alt.value(60))
).interactive()

scatter_chart
```

5.3. Используя данные об автомобилях, создайте двухпанельную (*two-panel*) гистограмму (количество миль на галлон на одной панели, количество лошадиных сил на другой), где вы можете перетащить мышь, чтобы выбрать данные на левой панели, чтобы отфильтровать данные на второй панели.

```python
brush = alt.selection_interval(encodings=['x'])

mpg_hist = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Miles_per_Gallon:Q', bin=True),
    y='count()',
).add_selection(
    brush
)
hp_hist = alt.Chart(cars).mark_bar().encode(
    x=alt.X('Horsepower:Q', bin=True),
    y='count()',
).transform_filter(
    brush
)

two_panel_hist = alt.hconcat(mpg_hist, hp_hist)
two_panel_hist.display()
```

5.5. Создайте диаграмму на основе самостоятельно сгенерированных данных и постройте кривые синуса и косинуса с помощью `transform_calculate`.

x = pd.DataFrame({'x': np.linspace(-len(Имя), len(Фамилия))})

```python
x = pd.DataFrame({'x': np.linspace(-len("Галина"), len("Сачкова"), 100)})

chart = alt.Chart(x).transform_calculate(
    sin='sin(datum.x)',
    cos='cos(datum.x)'
).mark_line().encode(
    x='x:Q',
    y='sin:Q',
    color=alt.value('blue')
).properties(
    title='Синус и Косинус'
) + alt.Chart(x).transform_calculate(
    sin='sin(datum.x)',
    cos='cos(datum.x)'
).mark_line().encode(
    x='x:Q',
    y='cos:Q',
    color=alt.value('red')
)

chart.display()
```

5.6. Используйте `transform_filter` на диаграмме задания 5 и удалите области графика, где значение кривой косинуса меньше значения кривой синуса.
**К сожалению, у меня точки разрыва соединяются прямой линией. Не смогла исправить это**

```python
 base_chart = alt.Chart(x).mark_line().encode(
    x='x:Q'
).transform_calculate(
    sin='sin(datum.x)',
    cos='cos(datum.x)'
)
sin_curve = base_chart.encode(
    y='sin:Q',
    color=alt.value('blue')
)
cos_curve = base_chart.transform_filter(
    (alt.datum.cos <= alt.datum.sin)
).encode(
    y='cos:Q'
)

filtered_chart = sin_curve + cos_curve
filtered_chart
```
